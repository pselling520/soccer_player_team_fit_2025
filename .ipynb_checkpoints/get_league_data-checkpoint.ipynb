```python
api_key = os.getenv("x-apisports-key")
```

```python
url = "https://v3.football.api-sports.io/leagues?country=USA"

payload={}
headers = {
  'x-apisports-key': api_key,
}

response = requests.request("GET", url, headers=headers, data=payload)

print(response.text)
```

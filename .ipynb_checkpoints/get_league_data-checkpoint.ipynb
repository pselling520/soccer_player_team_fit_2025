```python
import pandas as pd
import numpy as np
import requests
import ast
import json

from dotenv import load_dotenv
import os
```

```python
load_dotenv()
```

```python
api_key = os.getenv("x-apisports-key")
```

```python
url = "https://v3.football.api-sports.io/leagues?country=USA"

payload={}
headers = {
  'x-apisports-key': api_key,
}

response = requests.request("GET", url, headers=headers, data=payload)

print(response.text)
```

```python
usa_league_data = json.loads(response.text)
```

```python
mls_league_metadata = [year for year in usa_league_data['response'] if year['league']['name'] == 'Major League Soccer']
```
